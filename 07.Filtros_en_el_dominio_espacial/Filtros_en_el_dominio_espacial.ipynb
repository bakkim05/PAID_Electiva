```sos
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from scipy import signal
from scipy import special

A = Image.open('images/baby_yoda.jpg')
A_arr = np.asarray(A).astype('float64')
m,n = A_arr.shape

k = 5;


B = np.ones((k,k))
v = 0
for j in range(0,k):
    v = v+special.binom(k-1,j)

for i in range(0,k):
    for j in range(0,k):
        B[i,j] = 1/(v**2)*special.binom(k-1,j)*special.binom(k-1,i)*B[i,j]

for i in range(0,20):
    A_arr = signal.convolve2d(A_arr, B,mode = 'same')        
    
    
C_arr = np.clip(A_arr, 0, 255)
C = Image.fromarray(C_arr.astype('uint8'))

images =[]
fig, axs = plt.subplots(1, 2)
axs[0].set_title('Imagen Original')
axs[1].set_title('Filtro Gaussiano')
axs[0].set_xticks([])
axs[1].set_xticks([])
axs[0].set_yticks([])
axs[1].set_yticks([])
images.append(axs[0].imshow(A,cmap = 'gray'))
images.append(axs[1].imshow(C,cmap = 'gray'))
```
