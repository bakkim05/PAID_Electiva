```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow
```

```python
A = Image.open('images/barbara.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype('float64')/255
m,n,r = A_arr.shape

images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A_arr))

deltax = 20
deltay = 30
B_arr = np.zeros((m,n,r))
for x in range(0,m):
    for y in range(0,n):
        x_t = x + deltax
        y_t = y + deltay
        if (x_t < m) & (y_t < n):
            B_arr[x_t,y_t,:] = A_arr[x,y,:]

images.append(axs[1].imshow(B_arr))
```

```python
A = Image.open('images/barbara.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype('float64')/255
m,n,r = A_arr.shape

images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A_arr))
B_arr = np.zeros((m,n,r))
    
for i in range(0,min(m,n)):
    for x in range(0,m):
        for y in range(0,n):
            x_t = x + i
            y_t = y + i
            if (x_t < m) & (y_t < n):
                B_arr[x_t,y_t,:] = A_arr[x,y,:]

images.append(axs[1].imshow(B_arr))
```

```python
A = Image.open('images/boat.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype('float64')/255
m,n = A_arr.shape

B = np.zeros((m,n))

Am = 15
L = 75

for x in range(m):
    for y in range(n):
        xp = int(np.mod(round(x+1 + Am*np.sin(2*np.pi*(y+1)/L)),m))
        yp = int(np.mod(round(y+1 + Am*np.sin(2*np.pi*(x+1)/L)),n))
        B[xp, yp] = A_arr[x,y]
        
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
images.append(axs[1].imshow(B,cmap = 'gray'))
```
