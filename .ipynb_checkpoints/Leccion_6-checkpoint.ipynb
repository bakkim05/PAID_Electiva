```sos
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np

A = Image.open('Imagenes 4/boat.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype('float64')/255
m,n = A_arr.shape

B = np.zeros((m,n))

Am = 15
L = 75

for x in range(m):
    for y in range(n):
        xp = int(np.mod(round(x+1 + Am*np.sin(2*np.pi*(y+1)/L)),m))
        yp = int(np.mod(round(y+1 + Am*np.sin(2*np.pi*(x+1)/L)),n))
        B[xp, yp] = A_arr[x,y]
        
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
images.append(axs[1].imshow(B,cmap = 'gray'))
```

```sos
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np

A = Image.open('Imagenes 4/boat.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype('float64')/255
m,n = A_arr.shape

B = np.zeros((m,n))

L = 75
for Am in range(5,200,5):
    for x in range(m):
        for y in range(n):
            
            xp = int(np.mod(round(x+1 + Am*np.sin(2*np.pi*(y+1)/L)),m))
            yp = int(np.mod(round(y+1 + Am*np.sin(2*np.pi*(x+1)/L)),n))
            B[xp, yp] = A_arr[x,y]
        
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
images.append(axs[1].imshow(B,cmap = 'gray'))
```

```sos
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np

A = Image.open('Imagenes 4/boat.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype(np.int64)
m,n = A_arr.shape

b = 0
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
for c in np.arange(25,255,25):
    B_array = c*A_arr + b
    np.clip(B_array, 0, 255, out=B_array)
    B = Image.fromarray(B_array.astype('uint8'))
    images.append(axs[1].imshow(B,cmap = 'gray'))
```

```sos
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np

A = Image.open('Imagenes 4/boat_new.jpg')
A = A.resize((256,256))
A_arr = np.asarray(A).astype('uint8')
m,n = A_arr.shape
rmin = np.amin(A_arr)
rmax = np.amax(A_arr)

c = 255/(rmax-rmin)
b = -c*rmin

B_array = c*A_arr+b

B_array = np.clip(B_array, 0, 255)
B = Image.fromarray(B_array.astype('uint8'))
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
images.append(axs[1].imshow(B,cmap = 'gray'))

```
