```sos
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow

A = Image.open('Leccion 1/barbara.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype('float64')/255
m,n,r = A_arr.shape

images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A_arr))

deltax = 20
deltay = 30
B_arr = np.zeros((m,n,r))
for x in range(0,m):
    for y in range(0,n):
        x_t = x + deltax
        y_t = y + deltay
        if (x_t < m) & (y_t < n):
            B_arr[x_t,y_t,:] = A_arr[x,y,:]

images.append(axs[1].imshow(B_arr))
```

```sos
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow

A = Image.open('Leccion 1/barbara.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A).astype('float64')/255
m,n,r = A_arr.shape

images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A_arr))
B_arr = np.zeros((m,n,r))
    
for i in range(0,min(m,n)):
    for x in range(0,m):
        for y in range(0,n):
            x_t = x + i
            y_t = y + i
            if (x_t < m) & (y_t < n):
                B_arr[x_t,y_t,:] = A_arr[x,y,:]

images.append(axs[1].imshow(B_arr))
```

```sos
clc; clear; close all

pkg load image

A = imread('Leccion 1/barbara.jpg');
A = imresize(A, [128 128]);
subplot(2,2,1)
imshow(A)
title('Imagen Original')

[m,n,r] = size(A);
A_aux = uint8(zeros(floor(sqrt(m*m+n*n)),floor(sqrt(m*m+n*n)),r)); %matriz cuadrada con tamao igual a la hipotenusa
[m_aux,n_aux,r] = size(A_aux);
B = uint8(zeros(m_aux,n_aux,r)); %matriz para imagen rotada
B_r = uint8(zeros(m,n,r));% matriz para imagen recortada
xc = floor(m/2);
yc = floor(n/2);
xc_aux = floor(m_aux/2);
yc_aux = floor(n_aux/2);
deltax = xc_aux-xc;
deltay = yc_aux-yc;

%traslacion

for x = 1:m
    for y = 1:n
        x_t=x+deltax;
        y_t=y+deltay;
        A_aux(x_t,y_t,:) = A(x,y,:);
    end
end
subplot(2,2,2)
imshow(A_aux)
title('Imagen Escalada')

%rotacion

angulo = 32;
a0 = cosd(angulo);
a1 = sind(angulo);
b0 = -sind(angulo);
b1 = cosd(angulo);

for x = 1:m_aux
    for y = 1:n_aux
        x_t = mod(round(a0*(x-xc_aux)+a1*(y-yc_aux)+xc_aux),m_aux); %modulo m redondeado
        y_t = mod(round(b0*(x-xc_aux)+b1*(y-yc_aux)+yc_aux),n_aux); %modulo n redondeado
        B(x_t+1,y_t+1,:) = A_aux(x,y,:);
    end
end

subplot(2,2,3)
imshow(B)
title('Imagen Rotada')

%recorte

for x = 1:m_aux
    for y = 1:n_aux
        x_t=x-deltax;
        y_t=y-deltay;
        if and(x_t<=m,y_t<=n,x_t>0,y_t>0)
            B_r(x_t,y_t,:) = B(x,y,:);
        end
    end
end
subplot(2,2,4)
imshow(B_r)
title('Imagen Rotada Sin Exceso')
```
