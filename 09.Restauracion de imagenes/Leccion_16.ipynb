```sos
clc; clear; close all

pkg load image
A=imread('Images/camarografo.jpg');
subplot(1,3,1)
imshow(A)
title('Original')

N=imread('Images/ruido_periodico.jpg');
subplot(1,3,2)
imshow(N)
title('Ruido Periodico')

A = im2double(A);
N = im2double(N);

C=A+0.5*N;
C=im2uint8(C);
subplot(1,3,3)
imshow(C)
title('Ruidosa')
```

```sos
clc; clear; close all

pkg load image
A=imread('Images/camarografo.jpg');
N=imread('Images/ruido_periodico.jpg');
subplot(2,2,1)
imshow(A)
title('Original')

A = im2double(A);
N = im2double(N);

dA = fftshift(fft2(A));
subplot(2,2,2)
imshow(log(1+abs(dA)),[])
title('Original DFT-2D')

C=A+0.5*N;
dC = fftshift(fft2(C));
C=im2uint8(C);
subplot(2,2,3)
imshow(C)
title('Ruidosa')


subplot(2,2,4)
imshow(log(1+abs(dC)),[])
title('Ruidosa DFT-2D')

```
