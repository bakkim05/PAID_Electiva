```octave
%Ejemplo de una imagen a color

A = imread("imagenes/rgb_imagen.png");
A = double(A);
subplot(2,2,1)
imshow(A)
title("Imagen a Colores")

Ar = A(:,:,1); % Canal Rojo
subplot(2,2,2)
imshow(Ar)
title("Componente Rojo")

Ag = A(:,:,2); % Canal Verde
subplot(2,2,3)
imshow(Ag)
title("Componente Verde")

Ab = A(:,:,3); % Canal Azul
subplot(2,2,4)
imshow(Ab)
title("Componente Azul")
```

```octave
A = imread("imagenes/rgb_imagen.png");
A = im2double(A);
[m,n,r] = size(A);
subplot(2,2,1)
imshow(A)
title("Imagen a Colores")
Ceros = zeros(m,n);

%Canal Rojo Activado
A_r=zeros(m,n,3);
A_r(:,:,1) = A(:,:,1); % Canal Rojo
A_r(:,:,2) = Ceros;
A_r(:,:,3) = Ceros;
subplot(2,2,2)
imshow(A_r)
title("Componente Rojo Activado")

%Canal Verde Activado
A_g=zeros(m,n,3);
A_g(:,:,2) = A(:,:,2); % Canal Verde
A_g(:,:,1) = Ceros;
A_g(:,:,3) = Ceros;
subplot(2,2,3)
imshow(A_g)
title("Componente Verde Activado")

%Canal Azul Activado
A_b=zeros(m,n,3);
A_b(:,:,3) = A(:,:,3); % Canal Azul
A_b(:,:,2) = Ceros;
A_b(:,:,1) = Ceros;
subplot(2,2,4)
imshow(A_b)
title("Componente Azul Activado")
```
