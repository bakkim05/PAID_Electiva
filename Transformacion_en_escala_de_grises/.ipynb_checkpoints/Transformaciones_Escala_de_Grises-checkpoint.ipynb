# Transformación en escala de grises

* Sea $A \in {\rm I\!R}^{m\times n}$

## Transformación negativa

$\tau (A(x,y)) = 255 - A(x,y) =- A(x,y) + 255$

## Transformación de potencia / Transformacion Gamma
$\tau (A(x,y)) = C\cdot [A(x,y)]^\gamma$

Donde C y $\gamma$ son constantes positivas.
En el caso particular C = 1, entonces la función $\gamma$ es estrechar los valors oscuros de una entrada y amplía los niveles más claros en la salida de la imagen 

![](images/tranf-potencia.png) 

```sos
pkg load image;

A = imread('images/boat.jpg');
subplot(1,2,1);
imshow(A)

A = double(A);
c= 0.5; gamma = 1.2;
% c>1 aclara, c<1 oscurece
% gamma>1 aclara, gama<1 oscurece
B = c*A.^gamma; % .^ Eleva puntualmente cada una de las entradas de A.
B = uint8(B);
subplot(1,2,2);
imshow(B)
```
