```octave
pkg load image

A = imread("images/boat_new.jpg");
[m,n] = size(A);
A = double(A);

subplot(1,2,1)
imshow(A)
title("Imagen Original")

rmax = max(max(A));
rmin = min(min(A));
c = (255/(rmax-rmin));
b = -c*rmin;
B= c*A+b;
B=uint8(B);


subplot(1,2,2)
imshow(B)
title("Imagen Auto-Contraste")
```

```octave
pkg load image

A = imread("images/boat.jpg");
subplot(1,2,1)
imshow(A)
title("Imagen Original")

A = double(A);

%C>1, aclara; c<1 Oscurece
%gamma>1, aclara; gamma<1 Oscurece
c = 0.5; gamma = 1.2;

B = c*A.^gamma; % .^ eleva puntualmente cada una de las entradas de A.
B = uint8(B);

subplot(1,2,2)
imshow(B)
title("Imagen con Filtro Gamma")
```

TAREA: Transformada Logaritmica T(A(x,y)) = c * ln(1+A(x,y))

```octave
pkg load image

A = imread("images/boat.jpg");
subplot(1,2,1)
imshow(A)
title("Imagen Original")

A = double(A);

%C>1, aclara; c<1 Oscurece

c = 5;

B = c*log(1+A);
B = uint8(B);

subplot(1,2,2)
imshow(B)
title("Imagen con Transformada Logaritmica")
```

Tarea - Aplicar a la imagen "boat.jpg" la siguiente transformada.
T(A(x,y)) = (255 si A(x,y) >= 127) ; (0 si A(x,y) < 127)

```octave
pkg load image

A = imread("images/boat.jpg");
subplot(1,2,1)
imshow(A)
title("Imagen Original")

A = double(A);
[m,n] = size(A);

B = zeros(m,n);

for x=1:m
    for y=1:n
        if(A(x,y) >= 127)
            B(x,y) = 255;
        else
            B(x,y) = 0;
        endif
    end
end

subplot(1,2,2)
imshow(B)
title("Imagen Binaria de Boat")
```
