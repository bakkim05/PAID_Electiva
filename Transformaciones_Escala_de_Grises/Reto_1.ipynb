```python
#Import Libraries

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
#function

def filtro_gamma(imagen, c, gamma):
    A = Image.open(imagen)
    A = np.array(A).astype(np.double)

    B = c*np.power(A,gamma)
    B = np.array(B).astype(np.uint8)
    plt.imshow(B,cmap=plt.cm.gray)
```

```python
filtro_gamma("images/camarografo.jpg",1,1)
```

```python
def graph(imagen, c, gamma):
    A = Image.open(imagen)
    A = np.array(A).astype(np.double)
    A = np.resize(A,np.size(A))
    
    #X = np.arange(0,255,1)
    
    Y = c*np.power(A,gamma)
    
    plt.plot(A,Y)
```

```python
graph("images/camarografo.jpg",1,2)
```
