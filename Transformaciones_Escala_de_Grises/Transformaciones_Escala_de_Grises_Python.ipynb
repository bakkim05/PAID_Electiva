```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import imshow
```

```python
A = Image.open('images/boat.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A)
m,n = A_arr.shape

c = 0.5
b = 20

B_array = c*A_arr + b
B = Image.fromarray(B_array)
        
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
images.append(axs[1].imshow(B,cmap = 'gray'))
```

```python
A = Image.open('images/boat.jpg')
A = A.resize((128,128))
A_arr = np.asarray(A)
m,n = A_arr.shape

b = 0
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
for c in np.arange(10,-0.5,-0.5):
    B_array = c*A_arr + b
    B = Image.fromarray(B_array)
    images.append(axs[1].imshow(B,cmap = 'gray'))
```

```python
A = Image.open('images/boat_new.jpg')
A = A.resize((256,256))
A_arr = np.asarray(A).astype('uint8')
m,n = A_arr.shape
rmin = np.amin(A_arr)
rmax = np.amax(A_arr)

c = 255/(rmax-rmin)
b = -c*rmin

B_array = c*A_arr+b

B_array = np.clip(B_array, 0, 255)
B = Image.fromarray(B_array.astype('uint8'))
images =[]
fig, axs = plt.subplots(1, 2)
images.append(axs[0].imshow(A,cmap = 'gray'))
images.append(axs[1].imshow(B,cmap = 'gray'))
```
